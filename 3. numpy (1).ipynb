# Import Numpy

```python
import numpy as np
import random
```

# Create numpy arrays

```python
zeros_array = np.zeros((2, 3))  # 2x3 array of zeros
ones_array = np.ones((3, 3))    # 3x3 array of ones
random_array = np.random.randint( low=0, high =1129 , size=(4,5))  # 4x5 array of random numbers
```

```python
print(random_array)
```

# Accessing and Slicing

```python
random_array[0][0]
```

```python
random_array[-1][-1]
```

```python
random_array[0:2][-1][0:2]
```

# Manipulation

```python
random_array
```

```python
np.insert(random_array,2,10)
```

```python
np.delete(random_array,2)
```

# Airthmetic

```python
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])
addition = array1 + array2
subtraction = array1 - array2
multiplication = array1 * array2
division = array1 / array2
```

# Aggregation functions

```python
print(np.sum(random_array))
print(np.mean(random_array))
print(np.median(random_array))
print(np.min(random_array))
print(np.max(random_array))
```

# Reshaping

```python
reshaped_array = random_array.reshape(5, 4)  # Reshape to a 5x4 array
```

# Transposing

```python
transposed_array = reshaped_array.T
```

# Broadcasting

```python
scalar = 2
result = reshaped_array + scalar
```
